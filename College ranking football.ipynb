```python
from bs4 import BeautifulSoup
import requests
```

```python
url= 'http://www.espn.com/college-football/rankings'
```

```python
html=requests.get(url)
```

```python
html.text
```

```python
soup= BeautifulSoup(html.text,'html.parser')
```

```python
print(soup)
```

```python
result=soup.findAll('table',{'class':'rankings has-team-logos'})
```

```python
print(result[0])
```

```python
#find the header row
head=result[0].find('thead')
print(head)
```

```python
#get the text info in each head cell in an array which is in square bracket
ths=head.findAll('th')
print(ths)
```

```python
#To get only the text
label = [hc.get_text() for hc in ths]
print(label)
```

```python
#find the body(this will be an array)
body=result[0].findAll('tbody')
```

```python
#to get the rankings of all the teams
for row in body[0].findAll('tr'):
    tds=row.findAll('td')
    data = [bc.get_text() for bc in tds]
    print(data)
```
